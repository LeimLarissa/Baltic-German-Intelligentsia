## Functions

```python
def find_NomSpace_E(year, society, sourcedir, short_mode=True):
    '''
    This function takes the (str)year and (str)society abbreviation as well as a file directory and gives out the number of overall errors in short_mode. 
    In long mode it also shows the erroneus lines that have spaces between letters and double space between words.
    '''
    
    with open(sourcedir, 'r') as fp:                                  #I open the file as read(!)
        file = fp.readlines()                                         #I copy all lines from the file in a list
    
    findings = []                                                     #an empty list where I collect the line numbers of the error findings
    
    for num, line in enumerate(file, 1):                              #I go through each line of the file-list while also counting the lines starting by 1 instead of 0
        if re.match('^\S\s\S\s.*  ', line):                           #I search for the regex marking a none-white space followed by a white space (repeated once to eliminate false positive)at the beginning of the line and two joined spaces at somepoint in the line 
            findings.append(num)                                      #Adding the line number to the list (helps to manually cross check and to count the lines detechted)
            if short_mode:                                            #If the parameter short_mode is True
                continue                                              #I jump directly to giving the numer of overall findings
            else:
                print(str(num) + ' Line: ' + line)                    #Otherwise, I print the erroneus line including line number to manually cross check 
    print(len(findings))                                              #I print out the overall length of the error list 
```

```python
def find_SpecSpace_E(year, society, sourcedir, short_mode=True):
    '''
    This function takes the (str)year and (str)society abbreviation as well as a file directory and gives out the number of overall errors in short_mode. 
    In long mode it also shows the erroneus lines that have only spaces between letters and no double spaces between words.
    '''
    
    with open(sourcedir, 'r') as fp:                                  #I open the file as read(!)
        file = fp.readlines()                                         #I copy all lines from the file in a list
    
    findings = []                                                     #an empty list where I collect the line numbers of the error findings
    
    for num, line in enumerate(file, 1):                              #I go through each line of the file-list while also counting the lines starting by 1 instead of 0
        if re.match('^\S\s\S\s', line):                               #I search for the regex marking a none-white space followed by a white space (repeated once to eliminate false positive) at the beginning of the line
            findings.append(num)                                      #Adding the line number to the list (helps to manually cross check and to count the lines detechted)
            if short_mode:                                            #If the parameter short_mode is True
                continue                                              #I jump directly to giving the numer of overall findings
            else:
                print(str(num) + ' Line: ' + line)                    #Otherwise, in long mode I print the erroneus line including line numbers to manually cross check 
    print(len(findings))                                              #I print out the overall length of the error list 
```

```python
def add_Space_E_Stats(year, society, sourcedir):
    '''
    This function takes the (str)year and (str)society abbreviation as well as a file directory 
    and gives out a dictionary with the year, society, number of normal space errors and number of special errors for the given case. 
    The dictionary can be used to be added to an external DataFrame. 
    '''
    
    with open(sourcedir, 'r') as fp:                                 #I open the file as read(!)
        file = fp.readlines()                                        #I copy all lines form the file in a list
    
    nom_space_findings = []                                          #An empty list to store the normal space errors
    spec_space_findings = []                                         #An empty list to store the special space errors           
    
    for num, line in enumerate(file, 1):                             #I loop through all lines while counting them startin from 1 
        if re.match('^\S\s\S\s.*  ', line):                          #I search for the regex for normal space errors (spaces between letters and double spaces between words)
            nom_space_findings.append(num)                           #I append the line number to the empty list
            
        if  re.match('^\S\s\S\s', line):                             #I search for the regex for special space errors (spaces between letters and no double spaces between words)
            spec_space_findings.append(num)                          #I append the line number to an empty list 
    
    return {'YEAR': year, 'SOCIETY': society, 'NUM_NomSpace_ERROR': len(nom_space_findings), 'NUM_SpecSpace_ERROR': len(spec_space_findings)}            #I return a dictionary with the length of the created lists as the values for the Error-keys 
                                                                 
```
