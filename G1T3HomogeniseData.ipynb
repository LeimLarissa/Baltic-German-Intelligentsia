```python
from geopy.geocoders import Nominatim
import csv
import pandas as pd
import folium
import pandas as pd
from folium.plugins import MarkerCluster
```

```python
%matplotlib inline
```

```python
df = pd.read_csv("Output/GEG/1873-GEG-Members-ORedit.csv")
df.head()
```

```python
df = df.reset_index()

geolocator = Nominatim(user_agent='myapplication')

df['Place_Coordinates']
```

```python
#https://stackoverflow.com/questions/37600711/pandas-split-column-into-multiple-columns-by-comma
df_new = pd.concat([df, df['Place_Coordinates'].str.split(',', expand=True)], axis=1)

#https://stackoverflow.com/questions/11346283/renaming-column-names-in-pandas
df_new = df_new.rename(columns={0: 'latitude', 1: 'longitude'})

df_new = df_new.dropna(subset=['Place_Coordinates'])
df_new['Place_Coordinates']
```

```python
df_new
```

```python
#Jan's Edmond example notebook 
world_map = folium.Map(tiles="cartodbpositron")

marker_cluster = MarkerCluster().add_to(world_map)

for i in range(len(df_new)):
    # Get the latitude and longitude values
    lat = df_new.iloc[i]['latitude']
    long = df_new.iloc[i]['longitude']
    
    # Set the radius for the circle marker
    radius = 5
    
    # Set the text for the popup on the circle marker
    popup_text = """Location : {}<br>
                    Person : {}<br>
                    Occupation : {}<br>
                    Membership Status : {}<br>"""
    popup_text = popup_text.format(df_new.iloc[i]['Place'],
                                   df_new.iloc[i]['Names'],
                                   df_new.iloc[i]['Occupation'],
                                   df_new.iloc[i]['1L_MS'])
    # Create the circle marker and add it to the marker cluster
    folium.CircleMarker(location=[lat, long], radius=radius, popup=popup_text, fill=True).add_to(marker_cluster)

# Display the map
world_map
```
